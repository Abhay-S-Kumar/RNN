```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("muthuj7/weather-dataset")

print("Path to dataset files:", path)
```

```python
import os
os.listdir(path)
```

```python
df=pd.read_csv(os.path.join(path,'weatherHistory.csv'))
```

```python
df.head()
```

```python
df.info()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], utc=True)
df = df.set_index('Formatted Date')
df = df.drop(['Loud Cover'], axis=1)

display(df.head())
df.info()
```

```python
df_monthly_avg = df[['Temperature (C)', 'Apparent Temperature (C)']].resample('M').mean()
display(df_monthly_avg.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

columns_to_scale = ['Temperature (C)', 'Apparent Temperature (C)']

df_monthly_avg[columns_to_scale] = scaler.fit_transform(df_monthly_avg[columns_to_scale])

display(df_monthly_avg.head())
```

```python
def create_sequences(data, n_steps):
    X, y = [], []
    for i in range(len(data) - n_steps):
        seq = data.iloc[i : i + n_steps][['Temperature (C)', 'Apparent Temperature (C)']].values
        label = data.iloc[i + n_steps]['Temperature (C)']
        X.append(seq)
        y.append(label)
    return np.array(X), np.array(y)

n_steps = 12

X, y = create_sequences(df_monthly_avg, n_steps)

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

```python
total_samples = len(X)

train_split = int(0.7 * total_samples)
val_split = int(0.15 * total_samples)
test_split = total_samples - train_split - val_split

X_train, y_train = X[:train_split], y[:train_split]
X_val, y_val = X[train_split:train_split + val_split], y[train_split:train_split + val_split]
X_test, y_test = X[train_split + val_split:], y[train_split + val_split:]

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_val shape: {X_val.shape}")
print(f"y_val shape: {y_val.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, SimpleRNN, Dropout, Dense

model = Sequential()

model.add(Input(shape=(n_steps, X_train.shape[2])))

model.add(SimpleRNN(units=64, activation='relu'))

model.add(Dropout(0.2))

model.add(Dense(units=1))

model.summary()
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error', metrics=['mean_absolute_error'])

print("Model compiled successfully.")
```

```python
history = model.fit(X_train, y_train, batch_size=32, epochs=70, validation_data=(X_val, y_val))
print("Model training completed.")
```

```python
import matplotlib.pyplot as plt


train_loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(1, len(train_loss) + 1)

plt.figure(figsize=(10, 6))
plt.plot(epochs, train_loss, 'b', label='Training Loss')
plt.plot(epochs, val_loss, 'r', label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
print("Loss curves plotted successfully.")
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

loss, mae = model.evaluate(X_test, y_test, verbose=0)

y_pred = model.predict(X_test)

rmse = np.sqrt(loss)

r2 = r2_score(y_test, y_pred)

print(f"Test Loss (MSE): {loss:.4f}")
print(f"Test MAE: {mae:.4f}")
print(f"Test RMSE: {rmse:.4f}")
print(f"Test R-squared: {r2:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.7, label='Predictions')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'r--', label='Perfect Prediction')
plt.xlabel('Actual Temperature (Scaled)')
plt.ylabel('Predicted Temperature (Scaled)')
plt.title('Model Predictions vs. Actuals on Test Set')
plt.legend()
plt.grid(True)
plt.show()
print("Predictions vs. Actuals plot generated successfully.")
```

```python



last_n_steps_data = df_monthly_avg.iloc[-n_steps:][['Temperature (C)', 'Apparent Temperature (C)']].values


current_input_sequence = last_n_steps_data.copy()

forecasts = []

# 3. Loop 7 times to predict the next 7 days
for _ in range(7):
    # a. Reshape the current input sequence to be compatible with the model's input shape
    # (1, n_steps, number_of_features)
    input_for_prediction = current_input_sequence.reshape(1, n_steps, current_input_sequence.shape[1])

    # b. Use model.predict() to get the next temperature prediction
    next_temp_scaled = model.predict(input_for_prediction, verbose=0)[0, 0]

    # c. Append the prediction to the forecasts list
    forecasts.append(next_temp_scaled)

    # d. Update the input sequence by removing the oldest data point and adding the new prediction.
    #    For 'Apparent Temperature (C)', use 0.0 as a placeholder, as the model only predicts 'Temperature (C)'.
    new_data_point = np.array([[next_temp_scaled, 0.0]]) # Placeholder for Apparent Temperature
    current_input_sequence = np.vstack((current_input_sequence[1:], new_data_point))

print(f"Generated {len(forecasts)} forecasts (scaled).")
```

```python

# Create a dummy array with 0.0 for 'Apparent Temperature (C)'
dummy_forecasts_array = np.array([[temp, 0.0] for temp in forecasts])

# Inverse transform only the 'Temperature (C)' column effectively
# We need to inverse transform the entire array and then extract the first column
inverse_transformed_forecasts = scaler.inverse_transform(dummy_forecasts_array)[:, 0]

print("Inverse transformation of forecasts completed.")
print(f"Forecasted temperatures (Celsius): {inverse_transformed_forecasts}")
```

```python


last_date_in_df = df_monthly_avg.index[-1]
forecast_dates = pd.date_range(start=last_date_in_df + pd.offsets.MonthBegin(1), periods=len(forecasts), freq='MS')

print("Forecast dates generated successfully.")
print(f"Next 7 forecast dates: {forecast_dates.tolist()}")
```

```python


# 6. Create a new Pandas DataFrame with the forecasted dates and inverse-transformed temperatures.
df_forecast = pd.DataFrame({'Temperature (C)': inverse_transformed_forecasts}, index=forecast_dates)
df_forecast.index.name = 'Formatted Date'

display(df_forecast.head())
print("Forecast DataFrame created successfully.")
```

```python

df_combined = pd.concat([df_monthly_avg['Temperature (C)'], df_forecast['Temperature (C)']])

display(df_combined.tail(10))
print("Historical and forecast data combined successfully.")
```

```python


# 8. Plot the original df_monthly_avg data and the forecasted data on the same graph.
plt.figure(figsize=(15, 7))

# Plot historical data
plt.plot(df_monthly_avg.index, df_monthly_avg['Temperature (C)'], label='Historical Monthly Average Temperature', color='blue')

# Plot forecasted data
plt.plot(df_forecast.index, df_forecast['Temperature (C)'], label='Forecasted Monthly Average Temperature', color='red', linestyle='--')

plt.title('Monthly Average Temperature: Historical vs. Forecasted')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.legend()
plt.grid(True)
plt.show()

print("Historical and forecasted temperatures plotted successfully.")
```
